```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as sla

t1 = -1
t2 = -2
N = 20
```

```python
def construct_Hamiltonian(N,t1, t2):
    H = np.zeros((N,N))
    
    for i in range(N):
        for j in range(N):
            if abs(i - j) == 1:
                if (i + j) % 4 == 1:
                    H[i][j] = t1
                elif (i + j) % 4 == 3:
                    H[i][j] = t2
    return H

def dict_create(key_list, value_list):
    if len(key_list) != len(value_list):
        print ("Mismatch between number of keys and number of values")
        return False
    else:
        dict = {}
        for i in range(len(key_list)):
            dict[key_list[i]] = value_list[i]
        return dict
```

```python
A = construct_Hamiltonian(N,t1, t2)
eigenvalues, eigenvectors = np.linalg.eigh(A)

eigendict = dict_create(eigenvalues, eigenvectors)
    

x = np.zeros(N)

for i in range(N):
    x[i] = i

%matplotlib inline

for i in range(len(eigendict)):
    
    fig, ax = plt.subplots()
    ax.plot(x, eigendict[eigenvalues[i]])
    ax.set_xlabel(str(eigenvalues[i]))
```
